```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
```

### Let's download and prepare our data:

```python
(train_x, train_y), (test_x, test_y) = fashion_mnist.load_data()
train_x, test_x = train_x / 255.0, test_x / 255.0
```

### We create a basic model:

```python
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dropout(0.2),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_x, train_y, epochs=20, batch_size=128, validation_split=0.2, verbose=1)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Accurancy on train data')
plt.plot(history.history['val_accuracy'], label='Accurancy on test data')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accurancy graph (Adam optimizer)')
plt.show()


plt.plot(history.history['loss'], label='Accurancy on train data')
plt.plot(history.history['val_loss'], label='Accurancy on test data')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss graph (Adam optimizer)')
plt.show()
```

### Let's increase the number of layers of our neural network and determine the accuracy and loss.

```python
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_x, train_y, epochs=20, batch_size=512, validation_split=0.2, verbose=1)
```

We can see that after increasing the number of layers, the accuracy of the neural network improved and eventually reached 91%.

```python
plt.plot(history.history['accuracy'], label='Accurancy on train data')
plt.plot(history.history['val_accuracy'], label='Accurancy on test data')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accurancy graph (Adam optimizer)')
plt.show()


plt.plot(history.history['loss'], label='Accurancy on train data')
plt.plot(history.history['val_loss'], label='Accurancy on test data')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss graph (Adam optimizer)')
plt.show()
```

### Let's reduce the number of neurons on each layer and batch size:

```python
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_x, train_y, epochs=20, batch_size=128, validation_split=0.2, verbose=1)
```

We can see that the accuracy indicators have deteriorated a little compared to previous calculations.

### Let's increase the number of epochs:

```python
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='sigmoid'),
    layers.Dropout(0.2),
    layers.Dense(64, activation='sigmoid'),
    layers.Dropout(0.2),
    layers.Dense(32, activation='sigmoid'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_x, train_y, epochs=60, batch_size=128, validation_split=0.2, verbose=1)
```

We see that as the number of epochs increases, the accuracy of our model slowly increases.

```python
plt.plot(history.history['accuracy'], label='Accurancy on train data')
plt.plot(history.history['val_accuracy'], label='Accurancy on test data')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accurancy graph (Adam optimizer)')
plt.show()

plt.plot(history.history['loss'], label='Accurancy on train data')
plt.plot(history.history['val_loss'], label='Accurancy on test data')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss graph (Adam optimizer)')
plt.show()
```

### In the last example, we will change the optimizer from 'adam' to 'rmsprop'. We will also increase the number of epochs and use the 'early stop' regularization mechanism to prevent overtraining.

```python
from tensorflow.keras.callbacks import EarlyStopping

model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='sigmoid'),
    layers.Dropout(0.2),
    layers.Dense(64, activation='sigmoid'),
    layers.Dropout(0.2),
    layers.Dense(32, activation='sigmoid'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_accuracy', patience=4, restore_best_weights=True)

history = model.fit(train_x, train_y, epochs=60, batch_size=128, validation_split=0.2, verbose=1, callbacks=[early_stopping])
```

The early-stopping mechanism worked, determining the exact accuracy before passing all epochs, but the performance is lower than that of the other optimizer.

```python
plt.plot(history.history['accuracy'], label='Accurancy on train data')
plt.plot(history.history['val_accuracy'], label='Accurancy on test data')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accurancy graph (Adam optimizer)')
plt.show()

plt.plot(history.history['loss'], label='Accurancy on train data')
plt.plot(history.history['val_loss'], label='Accurancy on test data')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss graph (Adam optimizer)')
plt.show()
```

### So, we obtained the best accuracy rates (91.34%) with the following parameters: optimizer adam, number of layers 5, batch size 512, number of epochs 20.
